# Relatório de Análise III

Neste exercício o objetivo é listar todos os imóveis que sejam residenciais;

```python
import pandas as pd
dados = pd.read_csv('dados/aluguel.csv', sep=';')
dados.head()
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Flat',
 'Casa de Vila',
 'Loft',
 'Chácara',
 'Sítio',
 'Studio']
residencial
```

```python
#listando só os imóveis residenciais
dados_residencial = dados[dados['Tipo'].isin(residencial)]
dados_residencial
```

```python
#verificando os tipos que restaram no DF
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
#verificando a quantidade de registros
dados_residencial.shape[0]
```

```python
#reconstruindo o index
dados_residencial.index = range(dados_residencial.shape[0])
dados_residencial
```

### Exportando a base de dados

```python
#removemos o index, pois como default ele vem como True
dados_residencial.to_csv('dados/aluguel_residencial.csv', sep=';', index=False)
```
